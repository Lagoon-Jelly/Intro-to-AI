```python
# Pytorch
# ADD URL LATER

# Import Pytorch
import torch
from matplotlib import pyplot as plt

# Set seed to get same result everytime
torch.manual_seed(0)

# Generate Data
x = (6*(torch.rand(32,1))).sort(dim=0)[0]
y_true = 100*x
y = y_true + ((50)*(torch.rand(y_true.shape)-0.5))

# Plot the data
plt.plot(x, y_true, linestyle='--', label = 'Market Value')
plt.scatter(x, y, label = 'Sales Price')
plt.title('House Price & Square Footage')
plt.xlabel('Square Footage (1000\'s)')
plt.ylabel('House Price ($1000\'s)')

# Model Neural Network
model = torch.nn.Sequential(torch.nn.Linear(1, 1))
criterion = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 1e-2)

# Use the data to find right model
# AKA, use neural network/do machine learning
for epoch in range(500):

  # Forward
  pred_y = model(x)

  # Loss
  loss = criterion(pred_y, y)

  # Zero gradiants, backwards pass, and weight update
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

# Test model on new data
new_var = (6*(torch.rand(32,1))).sort(dim=0)[0]
pred_y

# Plot predictions
plt.plot(new_var, pred_y.detach().numpy(),'--r',
         label='Market Value (Predicted)')
plt.scatter(new_var, pred_y.detach().numpy(), color = 'r',
            label = 'Sales Price Predicted)')

# Add Legend
plt.legend()
```
