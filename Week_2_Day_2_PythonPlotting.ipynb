```python
# Import numpy & plotting tools
import numpy as np
import matplotlib.pyplot as plt

# Make time variable
t = np.arange(0,2,0.01)

#print(t)

# Specify frequency of cosine
f = 1

# Make cosine
y = np.cos(2*np.pi*f*t)
plt.plot(t,y, label = 'Original')

#Print shape of t
print(t.shape[0])

# Add noise to y
z = np.random.normal(0,0.1,t.shape[0])
#plt.plot(t,z)
y_noisy = y+z
plt.plot(t,y_noisy, label = 'Noisy')

# Import scipy
from scipy import signal

#Make filter, apply filter, and run
b, a = signal.butter(5,0.1,'low')
y_filtered = signal.filtfilt(b,a,y_noisy)
plt.plot(t,y_filtered, label = 'Filtered' )
plt.legend()
```

```python
# Minimal version of code
# Import numpy and plotting tools

import numpy as np
import matplotlib.pyplot as plt

# Make time variable
time_length = 4
t = np.arange(0,time_length,0.01)

# Specify frequency of cosine in Hertz (Hz)
f = 1.5

# Make cosine
y = np.cos(2*np.pi*f*t)
plt.plot(t,y,label = 'Original')

# Threshold y to figure out when beats are happening
y_thresholded = 1*(y>0.75)
plt.plot(t,0.75*y_thresholded, label = 'Thresholded')

# Get derivitive of y_thresholded
y_deriv = np.diff(y_thresholded)
plt.plot(t[0:-1],y_deriv,label = 'Derivative')

# Find heart rate
num_of_bts = np.sum(y_deriv == 1)
print(num_of_bts)
print(num_of_bts/time_length)
print(60*num_of_bts/time_length)

# Add legend
plt.legend()
```
